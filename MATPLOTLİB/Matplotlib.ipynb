## SUBPLOT() METHOD

```python
# CREAT MULTÄ° PLOTS ON THE SAME FÄ°GURE USÄ°NG THE .subplot() method. this method takes 3 parameters:
# nrows = number of rows
# ncols = number of columns
# plot_number = which refers to a specific plot in the figure
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,"green")
plt.subplot(1,2,2)
plt.plot(y,x,"red")
```

## LEGEND

```python
fig = plt.figure(figsize(8,6), dpi=100)
ax=fig.add_axes[0,0,1,1]
ax.plot(x,x**2, "green", label= "x square")
ax.plot(x,x**3, "red", label="x cube" )
ax.legend()
```
